 # Univariate analysis
 a. Examine Distributions of Individual Variables:
        Plot histograms or bar charts to visualize the distribution of key variables such as age, BMI, mental health days (Menthlth), and physical health days (PhysHlth).


```python
### 1. Distribution of Age (Categorical Variable) ###


# Set the aesthetic style for the plots
sns.set(style="whitegrid")

# Create a figure for the age distribution plot
plt.figure(figsize=(10, 6))

# Plotting the histogram and KDE for Age distribution
sns.histplot(df['Age'], kde=True, bins=30, color='skyblue')

# Add titles and labels
plt.title('Age Distribution of Individuals', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

# Conduct Bivariate analysis:-
a. Explore Relationships with Heart Disease:
Compare the distribution of key variables (e.g., HighBP, HighChol, BMI) between individuals with and without heart disease.

```python
### 1. Comparing HighBP Distribution ###
# Calculate counts for each category
counts = df.groupby(['HighBP', 'HeartDiseaseorAttack']).size().unstack(fill_value=0)
# Create the figure and axes
plt.figure(figsize=(10, 5))
# Plot the bars for each category
counts.plot(kind='bar', stacked=False, color=['lightgreen', 'lightcoral'], ax=plt.gca())
# Set titles and labels
plt.title('High Blood Pressure Distribution by Heart Disease Status')
plt.xlabel('High Blood Pressure (0 = No, 1 = Yes)')
plt.ylabel('Count')
# Set x-ticks and labels
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'], rotation=0)
# Add legend
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
# Show the plot
plt.show()
```

```python
### 2. Comparing HighChol Distribution ###
# Calculate counts for each category
counts = df.groupby(['HighChol', 'HeartDiseaseorAttack']).size().unstack(fill_value=0)
# Create the figure and axes
plt.figure(figsize=(10, 5))
# Plot the bars for each category
counts.plot(kind='bar', stacked=False, color=['lightgreen', 'lightcoral'], ax=plt.gca())
# Set titles and labels
plt.title('High Cholesterol Distribution by Heart Disease Status')
plt.xlabel('High Cholesterol (0 = No, 1 = Yes)')
plt.ylabel('Count')
# Set x-ticks and labels
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'], rotation=0)
# Add legend
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
# Show the plot
plt.show()
```

```python
### 3. Comparing BMI Distribution ###
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='HeartDiseaseorAttack', y='BMI', palette=['lightcoral', 'lightgreen'])
plt.title('BMI Distribution by Heart Disease Status')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('BMI')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'], rotation=0)
plt.show()
```

```python
### 4. Compare Smoker Distribution ###
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Smoker', hue='HeartDiseaseorAttack', palette=['lightgreen', 'lightcoral'])
plt.title('Smoking Status Distribution by Heart Disease Status')
plt.xlabel('Smoker (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'], rotation=0)
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

b. Visualize Correlations Between Variables:
Use scatter plots or correlation matrices to explore relationships between continuous variables.

```python
# Set the aesthetic style of the plots
sns.set(style="whitegrid")

### 1. Scatter Plots to visualize relationships between continuous variables ###
plt.figure(figsize=(15, 10))

# Scatter plot for BMI vs Age
plt.subplot(2, 2, 1)
sns.scatterplot(data=df, x='Age', y='BMI', hue='HighBP', style='HighChol', palette='deep')
plt.title('BMI vs Age')
plt.xlabel('Age')
plt.ylabel('BMI')

# Scatter plot for BMI vs. Mental Health Days
plt.subplot(2, 2, 2)
plt.scatter(df['BMI'], df['MentHlth'], color='blue', alpha=0.6)
plt.title('BMI vs. Mental Health Days')
plt.xlabel('BMI')
plt.ylabel('Mental Health Days')

# Scatter plot for BMI vs. Physical Health Days
plt.subplot(2, 2, 3)
plt.scatter(df['BMI'], df['PhysHlth'], color='orange', alpha=0.6)
plt.title('BMI vs. Physical Health Days')
plt.xlabel('BMI')
plt.ylabel('Physical Health Days')

# Scatter plot for Mental Health Days vs. Physical Health Days
plt.subplot(2, 2, 4)
plt.scatter(df['MentHlth'], df['PhysHlth'], color='green', alpha=0.6)
plt.title('Mental Health Days vs. Physical Health Days')
plt.xlabel('Mental Health Days')
plt.ylabel('Physical Health Days')

plt.tight_layout(pad=3.0)  # Add padding between subplots
plt.show()

### 2. Correlation Matrix ###
# Select continuous variables for the correlation matrix
continuous_vars = ['BMI', 'MentHlth', 'PhysHlth', 'Age']
# Calculate the correlation matrix
corr_matrix = df[continuous_vars].corr()
# Plot the heatmap with updated color map
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='viridis', linewidths=0.5, vmin=-1, vmax=1) 
plt.title('Correlation Matrix of Continuous Variables')
plt.show()
```

c. Compare Heart Disease Across Demographic Groups:
Analyze the prevalence of heart disease across different demographic groups (e.g., age, gender, education) using bar charts or pivot tables.

```python
sns.set(style="whitegrid")

### 1. Analyze Heart Disease by Age Group ###
age_counts = df.groupby(['Age', 'HeartDiseaseorAttack']).size().unstack(fill_value=0)
age_counts.plot(kind='bar', stacked=True, color=['lightgreen', 'lightcoral'], figsize=(10, 5))
plt.title('Heart Disease Prevalence by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

```python
### 2. Analyze Heart Disease by Gender ###
sex_counts = df.groupby(['Sex', 'HeartDiseaseorAttack']).size().unstack(fill_value=0)
sex_counts.plot(kind='bar', stacked=True, color=['lightgreen', 'lightcoral'], figsize=(10, 5))
plt.title('Heart Disease Prevalence by Gender')
plt.xlabel('Gender (0 = Male, 1 = Female)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

```python
### 3. Analyze Heart Disease by Education Level ###
education_counts = df.groupby(['Education', 'HeartDiseaseorAttack']).size().unstack(fill_value=0)
education_counts.plot(kind='bar', stacked=True, color=['lightgreen', 'lightcoral'], figsize=(10, 5))
plt.title('Heart Disease Prevalence by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

```python
### 4. Compare Heart Disease Across Income Levels ###
income_counts = df.groupby(['Income', 'HeartDiseaseorAttack']).size().unstack(fill_value=0)

plt.figure(figsize=(10, 5))
income_counts.plot(kind='bar', stacked=False, color=['lightgreen', 'lightcoral'], ax=plt.gca())
plt.title('Heart Disease Prevalence Across Income Levels')
plt.xlabel('Income Level')
plt.ylabel('Count')
plt.xticks(ticks=range(len(income_counts)), labels=[f'Level {i}' for i in income_counts.index], rotation=0)
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

# Conclusion:-  

The analysis of the dataset reveals significant insights into the prevalence of heart disease and its correlation with various health indicators. Approximately 10% of individuals reported having heart disease or a heart attack, with a notable association between heart disease and conditions such as high blood pressure (67%) and high cholesterol (62%). 

Higher Body Mass Index (BMI) levels among those with heart disease suggest obesity plays a critical role in cardiovascular risk. Additionally, smokers accounted for nearly 50% of individuals with heart disease, highlighting the impact of smoking on heart health. Physical activity emerged as a protective factor, with 80% of those without heart disease reporting regular exercise, while those with heart disease experienced more days of poor mental health.

Gender disparities were evident, as men reported higher heart disease rates, and socioeconomic factors, including education and income, correlated with lower disease prevalence. The correlation matrix confirmed that higher BMI was positively associated with high blood pressure and high cholesterol. These findings underscore the need for public health initiatives focused on education, regular screenings, and promoting physical activity to mitigate heart disease risks and enhance overall population health.
